```python
#creating new DataFrame

df=pd.read_csv("train (2).csv",usecols=['Neighborhood','Exterior1st','Exterior2nd'])
df.head()
```

```python
#Finding Unique Values

for x in df.columns:
    print(x,':',len(df[x].unique()),'labels')
```

```python
#Technique For Multi Categorical Variables
#Most Frequent variables
df.Neighborhood.value_counts().sort_values(ascending=False).head(20)

```

```python
#List of most frequent Categorical Variables
top_10=[x for x in df.Neighborhood.value_counts().sort_values(ascending=False).head(10).index]
top_10
```

# Make Binary

```python
for label in top_10:
    df[label]=np.where(df['Neighborhood']==label,1,0)
df[['Neighborhood']+top_10]
```

# All selected variable on OneHotEncoding

```python
def top_x(df2,variable,top_x_labels):
    for label in top_x_labels:
        df2[variable+"_"+label]=np.where(df[variable]==label,1,0)
df=pd.read_csv("train (2).csv",usecols=['Neighborhood','Exterior1st','Exterior2nd'])

top_x(df,'Neighborhood',top_10)
df.head(10)
```
